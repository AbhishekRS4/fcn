```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
metrics_fcn8 = np.load('model_fcn8_100/metrics.npy').item()
metrics_fcn16 = np.load('model_fcn16_100/metrics.npy').item()
metrics_fcn32 = np.load('model_fcn32_100/metrics.npy').item()
```

```python
x_values = np.arange(1, len(metrics_fcn8['train_loss']) + 1)

plt.figure(figsize = (16, 8))
plt.title('loss vs epoch')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.plot(x_values, metrics_fcn8['train_loss'], 'b--', label = 'fcn8 training loss')
plt.plot(x_values, metrics_fcn8['valid_loss'], 'r--', label = 'fcn8 validation loss')

plt.plot(x_values, metrics_fcn16['train_loss'], 'g--', label = 'fcn16 training loss')
plt.plot(x_values, metrics_fcn16['valid_loss'], 'c--', label = 'fcn16 validation loss')

plt.plot(x_values, metrics_fcn32['train_loss'], 'm--', label = 'fcn32 training loss')
plt.plot(x_values, metrics_fcn32['valid_loss'], 'k--', label = 'fcn32 validation loss')

plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16, 8))
plt.title('accuracy vs epoch')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.plot(x_values, metrics_fcn8['valid_acc'], 'b--', label = 'fcn8 validation accuracy')
plt.plot(x_values, metrics_fcn16['valid_acc'], 'g--', label = 'fcn16 validation accuracy')
plt.plot(x_values, metrics_fcn32['valid_acc'], 'r--', label = 'fcn32 validation accuracy')

plt.legend()
plt.show()
```
